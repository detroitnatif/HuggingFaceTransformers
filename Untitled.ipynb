```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Fetch historical data for Bitcoin and S&P 500 over the last 5 years
btc_hist = yf.Ticker("BTC-USD").history(period="5y")
sp500_hist = yf.Ticker("^GSPC").history(period="5y")

# Set the plot background to black
plt.style.use('dark_background')

# Plot Bitcoin closing prices
plt.figure(figsize=(12, 7))
btc_line, = plt.plot(btc_hist.index, btc_hist['Close'], label='BTC-USD', color='blue', linewidth=1)

# Plot S&P 500 closing prices on a secondary y-axis
sp500_ax = plt.gca().twinx()
sp500_line, = sp500_ax.plot(sp500_hist.index, sp500_hist['Close'], label='S&P 500', color='yellow', linestyle='--', linewidth=1)

# Enhance plot
plt.title('Bitcoin (BTC-USD) vs. S&P 500 - Last 5 Years', color='white')
btc_line.axes.set_xlabel('Date', color='white')
btc_line.axes.set_ylabel('BTC Closing Price (USD)', color='blue')
sp500_ax.set_ylabel('S&P 500 Closing Price', color='yellow')
lines = [btc_line, sp500_line]
labels = [l.get_label() for l in lines]
plt.legend(lines, labels, loc='upper left')
plt.grid(True, color='gray')

plt.show()
```
